EXTRACTING ITS COORIDNTAES FOR DOCKING

```python
# Extract XYZ coordinates from ligand PDB
coords = []

with open("lig_custom.pdb") as f:
    for line in f:
        if line.startswith("HETATM") or line.startswith("ATOM"):
            x = float(line[30:38])
            y = float(line[38:46])
            z = float(line[46:54])
            coords.append([x, y, z])

import numpy as np
coords = np.array(coords)

# Compute center
center = coords.mean(axis=0)
print(f"📍 Center of ligand:")
print(f"    X: {center[0]:.2f}")
print(f"    Y: {center[1]:.2f}")
print(f"    Z: {center[2]:.2f}")
```

NOW AGAIN DOCK(lets hope it does dock this time!)

```python
!./gnina -r rec.pdb -l lig_custom.pdb \
  --center_x 10.64 --center_y -0.89 --center_z -1.27 \
  --size_x 20 --size_y 20 --size_z 20 \
  --seed 42 \
  -o docked_custom_boxed.sdf.gz
```

#CONCLUSION
In our GNINA docking study on BRAF V600E (PDB ID: 3OG7), Vemurafenib showed better docking performance than Dabrafenib due to multiple factors: (1) Vemurafenib is the co-crystallized ligand in 3OG7, so redocking favored its structure, while Dabrafenib was cross-docked; (2) Vemurafenib’s rigid, planar scaffold fits the binding pocket well, whereas Dabrafenib’s bulkier, flexible structure led to suboptimal poses; (3) Vemurafenib formed stronger interactions with key residues like Cys532 and Phe595; and (4) Dabrafenib’s higher conformational flexibility reduced docking accuracy in the rigid receptor conformation.
